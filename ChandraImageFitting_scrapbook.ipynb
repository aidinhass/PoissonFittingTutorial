Note that the maximum of the data is $10^{-6}$. This'll cause problems in the Poisson likelihood!
Let's fix this by multiplying the smallest non-zero number to the data (maybe), and assume that that's one count.

```python
sb_nozero = sb_data[np.where(sb_data > 0.0)]
print(np.min(sb_nozero))
min_sbdata = np.min(sb_nozero)
```

Now we're going to assume that this number constitutes one photon: multiply data by this factor.

```python
## The ellipse model
def ellipse_model(x, cp_x, cp_y, q, a, phi, logScool, rcool, betacool, logbackground):

    q = np.exp(q)
    Scool = np.exp(logScool)
    background = np.exp(logbackground)

    theta = np.arctan2(x[1] - cp_x , x[0] - cp_y)
    radius = np.sqrt((x[0]-cp_x)**2 + ((x[1]-cp_y)/q)**2) * ( 1 + (a * np.cos(theta + phi)))
    profile = Scool*(1 + (radius/rcool) **2.)**(0.5 - 3*betacool) + background

    return profile
```

```python
logmin = 1000000000000.

class PoissonPosterior(object):
    ### x,y = Data
    ### func = physical model
    def __init__(self, x, y, func):
        self.x = x
        self.y = y

        ### func is a parametric model
        self.func = func

        return
    
    #def logprior(self, t0):
        
        ## THIS IS JUST AN EXAMPLE! WE NEED TO CHANGE THIS SO IT 
        ## ACTUALLY DOES THE RIGHT THING
        
        ## say background is t0[0]
    #    bkg  = t0[0]
    #    pbkg = bkg > 0
        
        ## this parameter can span a large range, so we're going to pick
        ## from the log-parameter because picking in linear space puts 
        ## undue weight to large values of the parameter. This is particularly
        ## useful to do for so-called scale parameters (like normalisation parameters etc.)
        #alpha = np.log(t0[1])
        #alpha_limits = [-10., 10.]
        
        ## probability of alpha is 1 if alpha is in range alpha_limits, 0 otherwise
        #palpha = (alpha >= alpha_limits[0] and alpha <= alpha_limits[1])
        
        ## third parameter is known to be around 1, 
        ## we pick a normal distribution with mean 1. and variance 0.1
        #beta = t0[2]
        #pbeta = scipy.stats.norm.pdf(beta, 1.0, 0.1)

    
        ## assuming all parameters are independent, we can multiply together 
        ## the individual probabilities for all parameters:
        #lp = pbkg*palpha*pbeta
    #    lp = pbkg
    #    return np.log(lp)
    
    def loglikelihood(self, t0):
        ## compute model counts
        model_counts = self.func(self.x, *t0)

        ## compute Poisson likelihood
        loglike = -np.sum(model_counts) + np.sum(self.y*np.log(model_counts))\
                    -np.sum(scipy.special.gammaln(self.y + 1))

        ## deal with NaN and inf values of the log-likelihood
        ## if either of these is true, choose a ridiculously small value
        ## to make the model *really* unlikely, but not infinite
        if np.isnan(loglike):
            loglike = -logmin
        elif loglike == np.inf:
            loglike = -logmin

        #print("loglikelihood: " + str(loglike))
        return loglike

    def logposterior(self, t0):
        #print("t0: " + str(t0))
        #print("log prior: " + str(self.logprior(t0)))
        #print("log likelihood: " + str(self.loglikelihood(t0)))
        lpost = self.loglikelihood(t0) #+ self.logprior(t0)
        #print("log posterior: " + str(lpost))
        return lpost 
    
    def __call__(self, t0, neg=False):
        lpost = self.logposterior(t0) 

        ## the negative switch makes sure you can call both
        ## the log-likelihood, and the negative log-likelihood
        if neg == True:
            return -lpost
        else:
            return lpost
    

    
```

```python
class PoissonImagePosterior(PoissonPosterior, object):
    
    def logprior(self, t0):
        #print("I am in the right logprior")
        #x, cp_x, cp_y, q, a, phi, logScool, rcool, betacool, logbackground
        
        ## dumbest priors possible: set by the boundaries of the image.
        
        ## first two parameters are center pixels
        cpx = t0[0]
        cpy = t0[1]
        
        ## center pixel needs to be within the image
        p_cpx = (cpx >= self.x[0][0,0] and cpx <= self.x[0][-1,-1])
        if p_cpx:
            p_cpx = 1./(self.x[0][-1,-1] - self.x[0][0,0])
        else:
            p_cpx = 0.
        
        p_cpy = (cpy >= self.x[0][0,0] and cpy <= self.x[0][-1,-1])
        if p_cpy:
            p_cpy = 1./(self.x[0][-1,-1] - self.x[0][0,0])
        else:
            p_cpy = 0.
        #print("pixel priors: " + str(p_cpx) + ", " + str(p_cpy))
        
        ## ellipticity q
        q = t0[2]
        p_q = (q >= -10. and q <= 10.)
        if p_q:
            p_q = 1./20.
        else:
            p_q = 0.
        
        #print("ellipticity prior: " + str(p_q))

        ## displacement of the center
        a = t0[3]
        hyper_a_mean = 0.0
        hyper_a_var = 1.0
        adist = scipy.stats.norm(hyper_a_mean, hyper_a_var)
        p_a = adist.pdf(a)
        #p_a = 1.
        
        #print("displacement prior: " + str(p_a))
        
        ## angle:
        ## This is a dumb prior. I need to find a better prior for
        ## periodic boundary conditions
        phi = t0[4]
        phi_hyper_mean = 0.0
        phi_hyper_var = 1./(2*np.pi)
        phi_dist = scipy.stats.vonmises(phi_hyper_var, phi_hyper_mean)
        #p_phi = phi_dist.pdf(phi) ## I think phi doesn't actually do anything?
        p_phi = 1.
        
        #print("angle prior: " + str(p_phi))
        
        ## Surface brightness of the core:
        ## flat in log-space translates to 1/S prior in real space
        logscool = t0[5]
        p_logscool = (logscool >= -10.0 and logscool <= 10.0)
        #p_logscool = 1.
        if p_logscool:
            p_logscool = 1./(10. - (-10.))
        else:
            p_logscool = 0.
        #print("surface brightness prior: " + str(p_logscool))
        
        #x, cp_x, cp_y, q, a, phi, logScool, rcool, betacool, logbackground

        ## core radius:
        ## dumb prior: radius needs to be smaller than 2*the image size
        rcool = t0[6]
        p_rcool = (rcool > 20. and rcool <= 20.*self.x[0][-1,-1])
        if p_rcool:
            p_rcool = 1./(20.*self.x[0][-1,-1] - 20.)
        else:
            p_rcool = 0.
        #print("radius prior: " + str(p_rcool))
        #p_rcool = 1.
        
        ## beta parameter
        ## let's put an exponential prior on this:
        beta = t0[7]
        
        #hyper_beta = 3.0 ## hyper parameter for the exponential prior
        if beta < 1./6.0 or beta > 20.0:
            p_beta = 0.0
        else:
            #p_beta = (1./hyper_beta)*np.exp(-(beta-1./6.)/hyper_beta)
            p_beta = 1./(20.-1./6.)
        #p_beta = 1.
        
        #print("beta prior: " + str(p_beta))
        
        ## log background, same as core radius, but probably a dumb choice
        logbkg = t0[8]
        p_logbkg = (logbkg >= -10. and logbkg <= 10.)
        if p_logbkg:
            log_bkg = (10. - (-10.))
        else:
            p_logbkg = 0.
        #print("background prior: " + str(p_logbkg))
        #p_logbkg = 1.
        #p_cpx = 1.
        #p_cpy = 1.
        #p_q = 1.
        #p_a = 1.
        #p_phi = 1.
        #p_logscool = 1.
        #p_rcool = 1.
        #p_beta = 1.
        #p_logbkg = 1.
        lp = p_cpx*p_cpy*p_q*p_a*p_phi*p_logscool*p_rcool*p_beta*p_logbkg
        #print("logprior: " + str(lp))
        
        if lp > 0:
           return np.log(lp)
        else:
           return -logmin
            #return -np.inf
        
    def logposterior(self, t0):
        #print("t0: " + str(t0))
        lprior = self.logprior(t0)
        print("log prior: " + str(self.logprior(t0)))
        llike = self.loglikelihood(t0)
        print("log likelihood: " + str(self.loglikelihood(t0)))
        lpost = lprior + llike
        #print("log posterior: " + str(lpost))
        return lpost 
    
    def __call__(self, t0):
        lpost = self.logposterior(t0) 

        ## the negative switch makes sure you can call both
        ## the log-likelihood, and the negative log-likelihood
        return lpost
    

```
