## A quick Poisson fitting tutorial in python

Requires:
- numpy
- scipy
- matplotlib

Data from the Chandra X-ray Satellite comes as images. These images are photon counting data, that is each pixel records an integer number of photons. 
Data of this kind follows a Poisson distribution, that is, if there was no signal (i.e. only background noise), the resulting photons would follow a Poisson distribution with a single parameter (the background rate). This means that the value in each pixel $y$ is picked from the following distribution:

$$P(y|\lambda) = \frac{e^{-\lambda}\lambda^{y}}{y!}$$ ,

where $\lambda$ is the model parameter describing the expected count rate.

The function we are interested in is the *likelihood* of the data. The likelihood describes the probability of observing the data we've measured, conditioned on a *physical* model (your surface brightness model) and a *statistical* model (the Poisson distribution). The physical model describes what you expect your image to look like if there was no noise. The statistical model describes the uncertainty in the measurements because there is noise present. 
The likelihood function is the product of the probabilities of all pixels, which is the probability of observing a number of counts in a pixel, $y_i$ under the assumption of a model count rate in that same pixel, $m_i$, multiplied together for all $N$ pixels.

All in all, the Poisson likelihood for a given physical model $m(\mathbf{\theta})$, which depends on a set of $K$ parameters $\mathbf{\theta} = \{\theta_1, \theta_2, ... , \theta_k\}$ looks like this:

$$L(\mathbf{\theta}) = P(\mathbf{y}|\mathbf{\theta}, H) = \prod_{i=0}^N{(\frac{e^{-m_i(\mathbf{\theta})}m_i(\mathbf{\theta})^{y_i}}{y_i!})}$$

The "best fit" parameters of your physical models are those that maximise the likelihood. In practical terms, the likelihood is often very large, and it is more convenient to maximise the logarithm of the likelihood. Because it is also easier for numerical reasons to minimise a function instead of maximising it, in practice you want to minimise the -log-likelihood, i.e. this quantity:

$$ 
-\log{(L(\mathbf{\theta})} = \sum_{i=0}^{N}{(-m_i(\mathbf{\theta}) + y_i\log{(m_i(\mathbf{\theta}))} - \log{(y_i!)}    )}
$$

You can separate out the various sums, and the logarithm of $y_i!$ becomes a $\Gamma$-function, which is in `scipy.special`. 

Below is some simple code for the Poisson likelihood of a given model and a stupidly simple example. Because I don't have image data, below I've used a lightcurve, but the whole thing should generalise easily to 2D and a more complicated model.

First, let's make some fake data. For now, this data is just a flat light curve of 1000 seconds duration and a time resolution of 1 second, with a background count rate of 10 counts/s. 

```python
import numpy as np
import scipy.stats
import scipy.optimize

## step 1: make some fake data, just a flat light curve with a 
## background parameter of 10

# time array
times = np.arange(0,1000,1)

# data array, pick from a Poisson distribution with mean rate=10
counts = np.random.poisson(10, size=len(times))
```

Next, let's define the model for what the background should be. In our case, it's just a flat background with a single parameter that describes the background count rate (which, at this point, we pretend we don't know).

```python

## function that describes the model
## this could have more parameters if your model is more complex
def mymodel(times, rate_par):
    ## this can be as complex as you like
    model_counts = np.ones(len(times))*rate_par
    return model_counts
    
```

Now we've got some data and we've got a physical model ("only background counts of an unknown flux") we'd like to use to describe this data. Now we need to define the Poisson likelihood, i.e. our statistical model.

```python
## this is the positive log-likelihood
## you can derive this easily from the definition of the poisson distribution
## rate is the 
def poisson_loglike(counts, model_counts):
    ## now we can compute the log-likelihood:
    llike = -np.sum(model_counts) + np.sum(counts*np.log(model_counts))-np.sum(scipy.special.gammaln(counts + 1))
    return llike
```

Now we can put the two together and compute the thing we actually want to minimise. This is also often called the *deviance*, for unknown (to me) reasons:

```python
## model is the function that describes the physical model
## parameters is a list with parameters (one or more)
## times and counts are the data
def deviance(model, parameters, times, counts):
    model_counts = model(times, *parameters)
    return -2.*poisson_loglike(counts, model_counts)
```

The class structure sets this problem up in a convenient way for fitting. Below, we're just plotting some guesses for the likelihood function to see how it changes with different values for the parameter. As you should be able to see, it gets larger (less negative) as we get closer to the true value of the rate (i.e. 10).

```python
## define PoissonLikelihood object:
pl = PoissonPosterior(times, counts, mymodel)


## print the likelihood function for some guesses to see its behaviour:
guesses = [[5.], [7.], [10.], [15.], [20.]]
for g in guesses:
    print(pl(g, neg=False))

```

Now let's do some actual fitting using scipy.optimize:

```python
    
## define your fit method of choice, look at documentation of scipy.optimize for details
## let's use the BFGS algorithm for now, it's pretty good and stable, but you can use others
## note that the output will change according to the algorithm, so check the documentation
## for what your preferred algorithm returns
fitmethod = scipy.optimize.fmin_bfgs

    
## set neg=True for negative log-likelihood

neg = True

initial_guess = [5.]
## do the actual fit
popt, fopt, gopt, covar, func_calls, grad_calls, warnflag  = fitmethod(pl, initial_guess, args=(neg,), full_output=True)

## popt are the optimum parameter values
## fopt is the likelihood function at optimum
## gopt is the value of the minimum of the gradient of the likelihood function
## covar is the inverse Hessian matrix, can be used for error calculation
## func_calls: number of function calls made
## grad_calls: number of calls to the gradient

print("Optimum parameter values: " + str(popt))
print("Likelihood at optimimum parameter values: " + str(fopt))
```

The returned value `bopt` is what we call the inverse of the Hessian matrix. The Hessian matrix is the matrix of second derivatives of the likelihood function with respect to each parameter in the model. This is computed numerically by the algorithm doing the fitting. 
One can show that the inverse of this matrix, called the *covariance matrix*, describes the variances and covariances between parameters (i.e. how much they vary with themselves, and with each other parameter). This tells you something about the uncertainty in the parameters (via the variance) and how much they correlate  with each other (the covariances).
You can compute the standard errors on your parameters by taking the square-root of the diagonal of this matrix (the square root of the variances):


```python
stderr = np.sqrt(np.diag(covar))
print("standard error on the parameters: " + str(stderr))
```

In our stupid simple example, there's only one parameter, so only one element in the inverse Hessian matrix. If you have two parameters, this matrix will have four elements, for three parameters 9 elements and so forth.

## MCMC for this problem using *emcee*

Markov Chain Monte Carlo is a powerful technique to recover complex probability distributions. For example, imagine your negative log-likelihood function has two minima for a given parameter. In this case, the fitting algorithm may end up at one minimum or another, depending on the starting value, and you will never know about the other. Or, imagine that your errors are skewed: your estimate may be much more uncertain in one direction than another. In this case, your standard errors will not reflect reality very well.
In this case, it can be useful to *sample* the probability distribution instead of fit it in order to retain enough information about the distribution to make useful estimates. 

This is not an intro into MCMC, but there are many good tutorials out there. If you'd like to learn about MCMC, I'd recommend to write your own Metropolis-Hastings sampler. 

Below, I'll be using an out-of-the-box, very stable and well-written code called *emcee* to run MCMC on the likelihood above. This may take a while to run, depending on the number of data points and the complexity (number of parameters) of your model

```python
import emcee

## number of walkers in emcee
## should be large (>100)
nwalkers = 500

### number of dimensions for the Gaussian seeds (= number of parameters)
ndim = len(popt)

### sample random starting positions for each of the walkers
p0 = [np.random.multivariate_normal(popt,covar) for i in xrange(nwalkers)]

### initialize sampler
sampler = emcee.EnsembleSampler(nwalkers, ndim, pl, args=[False])
```

Now burn in the sampler:

```python
### run burn-in phase and reset sampler
burnin_samples = 200
pos, prob, state = sampler.run_mcmc(p0, burnin_samples)
sampler.reset()

```

Run actual MCMC chain:

```python


### run actual MCMCs
niter = 100
sampler.run_mcmc(pos, niter, rstate0=state)

### list of all samples stored in flatchain
mcall = sampler.flatchain

### print meanacceptance rate for all walkers and autocorrelation times
print("The ensemble acceptance rate is: " + str(np.mean(sampler.acceptance_fraction)))
L = np.mean(sampler.acceptance_fraction)*len(mcall)
acceptance = np.mean(sampler.acceptance_fraction)
#try:
#    self.acor = sampler.acor
#    print("The autocorrelation times are: " +  str(sampler.acor))
#except ImportError:
#    print("You can install acor: http://github.com/dfm/acor")
#    self.acor = None
#except RuntimeError:
#    print("D was negative. No clue why that's the case! Not computing autocorrelation time ...")
#    self.acor = None
#except:
#    print("Autocorrelation time calculation failed due to an unknown error: " + sys.exc_info()[0] + ". Not computing autocorrelation time.")
#    self.acor = None

```

For multi-parameter problems, you can make a matrix plot that plots scatter plots for each parameter against each other parameter, and histograms for each parameter individually. This is a tremendously useful diagnostic tool in determining whether your distributions are skewed, or whether there are significant correlations between parameters. 

Here's code for the matrix plot. It makes a simple scatter plot like the one above in the case of a one-parameter model, and a matrix plot for a model with several parameters.

```python

## function that makes the scatter plot
def scatter_plot(samples, ndims):
    fig, axes = plt.subplots(ndims,ndims,figsize=(15,15))

    ### for one-parameter model, make scatter plot
    if ndims == 1:
        axes.hist(samples, bins=20)
        
    ### for more than one parameter, make matrix plot
    else:
        for i in xrange(ndims): ## x dimension
            for j in xrange(ndims): ## y dimension
                if i == j: 
                    axes[i,j].hist(samples[:,i], bins=20)
                else:
                    axes[i,j].scatter(samples[:,i], samples[:,j], color="black")

    return


## number of dimensions for the plot = number of parameters
ndims = mcall.shape[1]

## number of MCMC samples
nsamples = mcall.shape[0]

## make scatter plot for real model:
scatter_plot(mcall, ndims)


```

You can see that some parameters have strong correlations. The histograms all look quite Gaussian, but this is because I put in Gaussian distributions to simulate the samples.

The advantage of MCMC is that you can easily make simulations from the samples you drew from the model. This can be quite useful when you'd like to, for example, make sure the errors in your parameter estimation are properly accounted for in whatever final result you're interested in (I guess in your case, the cavity regions in the image). 
The basic procedure works like this:
- randomly pick a parameter set from your MCMC sample
- make an image from that model
- for each pixel in this image, pick from a Poisson distribution where the distribution's parameter $\lambda$ is the pixel value derived from your model. We're going to pretend that this is data (without cavity)
- run whatever cavity detection procedure you use on this simulated image and compute whatever final result you're interested in
- repeat these four steps many (say, 1000$ times
- compare the numbers derived from your simulations to that from your real observed image. If the numbers are very similar, you probably don't have a cavity. If they are very different, either there's something real there or your background model is wrong.

```python

## This is Martijn's ellipse model
def ellipse_model(x, cp_x, cp_y, q, a, phi, logScool, rcool, betacool, logbackground):

    q = np.exp(q)
    Scool = np.exp(logScool)
    background = np.exp(logbackground)
    
    theta = np.arctan2(x[1] - cp_x , x[0] - cp_y)
    radius = np.sqrt((x[0]-cp_x)**2 + ((x[1]-cp_y)/q)**2) * ( 1 + (a * np.cos(theta)+ phi))
    profile = Scool*(1 + (radius/rcool) **2.)**(0.5 - 3*betacool) + background

    return profile
```

### Defining a Posterior Distribution

Here comes the fun bit. In order to do MCMC correctly, we need to define something called a *posterior distribution*. To explain that, I'll need to explain some statistical concepts first, mostly something called **Bayesian statistics**. 

In "classical" statistics, a probability is defined as a *frequency of events*. That is, for example, the probability of getting "heads" on a coin flip is defined as the number a coin lands on "heads" for $N$ coin tosses, where $N$ goes to infinity. This definition holds true for truly probabilistic events only (e.g. coin tosses, dice rolls, the chance of getting a photon in a detector, etc). It does not hold for variables that aren't intrinsically random, say, the mass of a star or galaxy. 

Bayesian statistics, on the other hand, defines probability as a *degree of belief* in a variable. It is still possible to define the probability of a coin landing on "heads", but the probability is nor rather defined as my *degree of belief* in getting heads in the next coin toss, rather than a frequency defined through a million experiments (although Bayesian statistics will also say that the more experiments you make, the more precise your degree of belief will be). 
While this may, in the first instance, seem really counter-intuitive, this definition is actually closer to how we use the word probability in real life. When we talk about the "probability that it will rain tomorrow", we don't talk about the hundreds of experiments we've made (we haven't), but our degree of belief in the fact that it'll rain tomorrow. Also, the weather is not a random variable, thus defining a classical probability would be impossible.

What does this all have to do with X-ray data and galaxy clusters? Well, one advantage that Bayesian statistics has over classical statistics is that it obeys probability theory, and can do a lot of stuff that classical statistics can't do without some weird contortions. 
More specifically, when we talk about a likelihood function, we always talk about the *probability of observing the data, given a model*. Formally, this is defined in the following way:

$$P(y|\lambda) = \frac{e^{-\lambda}\lambda^{y}}{y!}$$ .

If you look closely, this is the equation I defined all the way at the top in the first paragraph. This is the probability of getting a data set ($y$, here your image of the galaxy cluster) given a model (here, the model is that the data follows a Poisson distribution with rate parameter $\lambda$). 

For your elliptical galaxy model, we've defined a slightly more complex probability distribution above.

$$L(\mathbf{\theta}) = P(\mathbf{y}|\mathbf{\theta}, H) = \prod_{i=0}^N{(\frac{e^{-m_i(\mathbf{\theta})}m_i(\mathbf{\theta})^{y_i}}{y_i!})}$$

We now have two models! A physical model (the elliptical galaxy model) and a statistical one (Poisson distribution). But you'll notice here that we're calculating something we're not technically interested in! We can now use a fitting algorithm to find the parameter set where this function has its maximum, but we will always compute the probability of observing the *data* given the model, never the probability of the model (the parameters) given the data! For the kind of things you're interested in, this might be important! Classical statistics can still kind of give you what you want (errors on your parameters), but it requires some awkwardness to get there.

Thanks to the rules of probability, Bayesian statistics can turn this equation around, by multiplying your likelihood by a *prior distribution*. This prior distribution is something you *choose* (and there's a whole subfield in statistics concerned with how to choose priors) and encapsulates everything you knew about the problem *before* you gathered your data. For example, if you'd previously measured the brightness of the galaxy, then this could go into the prior distribution. 

If you multiply the two together, you get the following:

$$P(\mathbf{\theta} | \mathbf{y}, H) = \frac{P(\mathbf{y}|\mathbf{\theta}, H) P(\mathbf{\theta} | \mathbf{y}, H)}{P(\mathbf{y} | H)}$$

Let's ignore the term in the denominator. This is the probability of the data, *given the model $H$ only*, i.e. integrating out the uncertainty in the parameters. This is incredibily hard to estimate, but luckily for us, it doesn't depend on the parameters, so for parameter estimation, we can ignore it without changing our results! Note however, that for model comparisons (say, if you had different galaxy models), this becomes important. 

The term in the nominator is the likelihood (first term) multiplied by the prior distribution (second term), which we haven't chosen yet.
You'll notice that the term on the left-hand side is now the thing we're actually interested in: the probability in the *parameters*, given the *data* and the *model*. This allows us to do several interesting things. For example, we can integrate over all parameters except one, and plot the distribution of that parameter alone (you can't do that with the likelihood, because you're not allowed to integrate over anything on the right side of the given ("|") bar!). 

So, below I've defined a posterior for your problem. The likelihood is the same as defined above already. For the prior, I've made some fairly generic choices the we can talk about at some point. The prior distribution defines the prior knowledge we have about the problem. I've chosen independent priors for each parameter (that is, we assume each parameter in the model is independent), and fairly generic ones within sensible bounds for the problem. The most important thing about priors is that they need to integrate to 1, thus cannot be infinite!

The implementation is a class, but that shouldn't put you off. I've put some code in to show you how to call the class and compute the posterior probability of a given parameter set.


```python
class PoissonImagePosterior(object):

    def __init__(self, x, y, func):
        """
        This method is required for every class. It defines what happens when you first 
        call the class, and which inputs to give it. 
        Here, give it the x-coordinate grid, the data, and a function (your elliptical function)
        to fit.
        
        >>> lpost = PoissonImagePosterior(xdata, sb_data, ellipse_model)
        
        You can then call the methods defined below in the following way, assuming you have
        some parameter set defined somewhere.
        
        >>> mylogprior = lpost.logprior(myparset)
        >>> myloglikelihood = lpost.loglikelihood(myparset)
        >>> mylogposterior = lpost.logposterior(myparset)
        >>> mylogposterior2 = lpost(myparset)
        >>> mynegativelogposterior = lpost(myparset, neg=True)
        
        See also the descriptions below for details
        """
        
        ## store x-coordinate grid in self.x and data in self.y
        ## if the object you've defined above is called "lpost", call either with 
        ## >>> lpost.x
        ## >>> lpost.y
        self.x = x
        self.y = y

        ### func is a parametric model, for example ellipse_model
        self.func = func

        return
    
    
    def loglikelihood(self, t0):
        
        ## call the function you've put in __init__, use parameter set t0 and
        ## the x-coordinate grid to compute the model image
        model_counts = self.func(self.x, *t0)

        ## compute Poisson likelihood
        loglike = -np.sum(model_counts) + np.sum(self.y*np.log(model_counts))\
                    -np.sum(scipy.special.gammaln(self.y + 1))

        ## deal with NaN and inf values of the log-likelihood
        ## if either of these is true, choose a ridiculously small value
        ## to make the model *really* unlikely, but not infinite

        if np.isnan(loglike):
            loglike = -logmin
        elif loglike == np.inf:
            loglike = -logmin

        return loglike

    
    def logprior(self, t0):
        """
        We're going to define prior distributions for all parameters in the 
        ellipse_model function. This means that this particular implementation is pretty
        inflexible when it comes to different models (i.e. if you had an alternative model to
        ellipse_model, but we can deal with that later if necessary at all.
        """
         
        #print("I am in the right logprior")
        #x, cp_x, cp_y, q, a, phi, logScool, rcool, betacool, logbackground
        
        ## dumbest priors possible: set by the boundaries of the image.
        
        ## first two parameters are center pixels
        cpx = t0[0]
        cpy = t0[1]
        
        ## center pixel needs to be within the image
        ## p_cpx will be True if cpx is between 0 and 801 (the image boundaries),
        ## or False otherwise
        p_cpx = (cpx >= self.x[0][0,0] and cpx <= self.x[0][-1,-1])
        
        ## turn probability into numbers: if p_cpx True, then the probability is 
        ## 1/(801-0), otherwise it's zero. This translates to a uniform (flat) 
        ## probability distribution on cpx, i.e. we have no clue where the center 
        ## pixel of the distribution should be, it could be anywhere on the x-axis!
        if p_cpx:
            p_cpx = 1./(self.x[0][-1,-1] - self.x[0][0,0])
        else:
            p_cpx = 0.
        
        ## Same for cpy as for cpx
        p_cpy = (cpy >= self.x[0][0,0] and cpy <= self.x[0][-1,-1])
        if p_cpy:
            p_cpy = 1./(self.x[0][-1,-1] - self.x[0][0,0])
        else:
            p_cpy = 0.
        #print("pixel priors: " + str(p_cpx) + ", " + str(p_cpy))
        
        ## ellipticity q
        ## I assumed some values within which the model kind of still looks like a galaxy,
        ## but chosen them pretty broad. Same reasoning as for cpx and cpy applies.
        ## Note that q is actually the *log* of the ellipticity!
        q = t0[2]
        p_q = (q >= -10. and q <= 10.)
        if p_q:
            p_q = 1./20.
        else:
            p_q = 0.
        
        #print("ellipticity prior: " + str(p_q))

        ## displacement of the center
        ## For the displacement of the center, I've chosen a normal distribution for the 
        ## prior, which puts a stronger preference on the model being circular than 
        ## being weirdly shaped. Again, this is basically a result of me playing around 
        ## with the model a lot!
        a = t0[3]
        hyper_a_mean = 0.0
        hyper_a_var = 1.0
        adist = scipy.stats.norm(hyper_a_mean, hyper_a_var)
        p_a = adist.pdf(a)
        #p_a = 1.
        
        #print("displacement prior: " + str(p_a))
        
        ## angle:
        ## This is a dumb prior. I need to find a better prior for
        ## periodic boundary conditions
        
        ## Above is a note for me. I'm not a hundred percent sure what to do 
        ## with a periodic parameter like an angle. So I've given it a von Mises 
        ## distribution, which translates to a normal distribution on a circle, but
        ## I need to check this carefully! For now, it works(ish)
        phi = t0[4]
        phi_hyper_mean = 0.0
        phi_hyper_var = 1./(2*np.pi)
        phi_dist = scipy.stats.vonmises(phi_hyper_var, phi_hyper_mean)
        #p_phi = phi_dist.pdf(phi) ## I think phi doesn't actually do anything?
        p_phi = 1.
        
        #print("angle prior: " + str(p_phi))
        
        ## Surface brightness of the core:
        ## flat in log-space translates to 1/S prior in real space
        
        ## Again, I've chosen sensible ranges for the logarithm of the parameter,
        ## and then applied a flat prior. Note that while the image only has integer counts,
        ## logscool doesn't have to be integer (it's a real physical quantity, and the fact
        ## that the counts are only integer comes from the Poisson distribution, which is a 
        ## property of the data!)
        logscool = t0[5]
        p_logscool = (logscool >= -10.0 and logscool <= 10.0)
        #p_logscool = 1.
        if p_logscool:
            p_logscool = 1./(10. - (-10.))
        else:
            p_logscool = 0.
        #print("surface brightness prior: " + str(p_logscool))
        
        
        ## core radius:
        ## dumb prior: radius needs to be smaller than 20*the image size
        
        ## I've noticed that rcool and beta correlate (I can make rcool smaller by
        ## increasing beta, or the other way around or something). So I gave 
        ## rcool a flat, very wide prior.
        rcool = t0[6]
        p_rcool = (rcool > 20. and rcool <= 20.*self.x[0][-1,-1])
        if p_rcool:
            p_rcool = 1./(20.*self.x[0][-1,-1] - 20.)
        else:
            p_rcool = 0.
        #print("radius prior: " + str(p_rcool))
        
        ## beta parameter
        ## let's put an exponential prior on this:
        beta = t0[7]

        ## I've noticed that when beta is < 1/6, the profile becomes inverted 
        ## (high values in the corners, low values in the center), so I've forced
        ## beta to be > 1/6, and otherwise flat up to a value of 20.
        if beta < 1./6.0 or beta > 20.0:
            p_beta = 0.0
        else:
            #p_beta = (1./hyper_beta)*np.exp(-(beta-1./6.)/hyper_beta)
            p_beta = 1./(20.-1./6.)
        #p_beta = 1.
        
        #print("beta prior: " + str(p_beta))
        
        ## log background, same as logScool
        logbkg = t0[8]
        p_logbkg = (logbkg >= -10. and logbkg <= 10.)
        if p_logbkg:
            log_bkg = (10. - (-10.))
        else:
            p_logbkg = 0.
        
        ## lots of commented out stuff for testing, ignore!
        #print("background prior: " + str(p_logbkg))
        #p_logbkg = 1.
        #p_cpx = 1.
        #p_cpy = 1.
        #p_q = 1.
        #p_a = 1.
        #p_phi = 1.
        #p_logscool = 1.
        #p_rcool = 1.
        #p_beta = 1.
        #p_logbkg = 1.
        lp = p_cpx*p_cpy*p_q*p_a*p_phi*p_logscool*p_rcool*p_beta*p_logbkg
        #print("logprior: " + str(lp))
        
        ## if the prior is larger than zero, return the log of it!
        ## if it's zero, return log of some miniscule nonzero value, since the 
        ## log of 0 is NaN!
        if lp > 0:
           return np.log(lp)
        else:
           return -logmin
            #return -np.inf    
        
    def logposterior(self, t0):
        """
        This computes the posterior distribution of the model parameters 
        given the data and the model (in our case the ellipse model).
        
        First, it computes the log-prior. Then it computes the log-likelihood,
        and in the last step, adds them all together.
        
        Since we're only interested in parameter estimation for now, we're ignoring
        the term in the denominator. 
        """
        
        lprior = self.logprior(t0) ## compute prior
        print("log prior: " + str(self.logprior(t0)))
        
        llike = self.loglikelihood(t0) ## compute likelihood
        print("log likelihood: " + str(self.loglikelihood(t0)))
        
        lpost = lprior + llike ## compute log-posterior
        #print("log posterior: " + str(lpost))
        
        return lpost 
    
    def __call__(self, t0, neg=False):
        """
        This function defines what happens when you simply do 
        >>> lpost(myparameterset)
        
        In this case, it computes the posterior, but it also 
        has a switch whether to return the *negative* posterior 
        (to use in minimization routines) or the positive posterior 
        (e.g. for MCMC).
        """

        lpost = self.logposterior(t0) 

        ## the negative switch makes sure you can call both
        ## the log-likelihood, and the negative log-likelihood
        if neg == True:
            return -lpost
        else:
            return lpost
    

```

Let's now run some diagnostics, the same plot we've done above.
First, check the ensemble acceptance rate above: I got 0.3725. This is quite good.
It shouldn't be too high: a too high acceptance rate means that nearly every sample gets accepted. While that might sound good in principle, it's not. A high acceptance rate means your chain effectively does a random walk through parameter space; unless your posterior distribution is flat (and that would be very, very, **very** bad, because then the data would tell you exactly **nothing**), it shouldn't be a random walk, so as long is it does a pure random walk, you're not approximating your posterior distribution, i.e. the thing you're interested in.

On the other hand, if your acceptance rate is super-small, that means very few samples get accepted. This is bad, too, because if only very few samples get accepted, there will be correlations between many neighbouring samples in the chain. This is bad, because the sample we ultimately want is a sample of *independent* parameter sets drawn from the posterior distribution. If there are correlations in the MCMC chain, this independence won't be true, so the distribution is again not the posterior we're interested in.

Let's now plot the different posterior distributions of parameters and how they correlate with each other. Note that we can only do this because we're using Bayesian statistics here!
I said earlier that one of the great advantages is that you can integrate out parameters you're not interested in! This is true because the MCMC sample gives you, in the ideal case, an approximation of the posterior distributions of your parameters given the data. Here, this is a 9-dimensional function (so pretty difficult to visualise). The cool thing you can do is integrate over any of those 9 dimensions to get rid off, and you'll have the posterior distribution for the rest.

In practice, this is super-easy. The posterior distribution of, say, the x-axis pixel position (the first parameter), you simply look at the first parameter of *all* samples from the MCMC chain, and you're done! You could also, for example, look at all x-axis positions where the y-axis position is, say 400. In this case, you'd look at the posterior for the x-pixel position, given that the y-axis position is 400 and the data, and integrated over all other parameters. Phew.
Sounds complicated, I know, but it really isn't. 
The plot below is super-useful for looking at the distributions of individual parameters, and how well they correlate with others.
